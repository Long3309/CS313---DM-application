```python
def apriori(transactions, min_support):
    # Tính số lần xuất hiện của các phần tử duy nhất trong các giao dịch
    item_counts = {}
    for transaction in transactions:
        for item in transaction:
            if item in item_counts:
                item_counts[item] += 1
            else:
                item_counts[item] = 1
    
    # Lọc các phần tử không đáp ứng ngưỡng hỗ trợ tối thiểu
    frequent_items = set(item for item in item_counts if item_counts[item]/len(transactions) >= min_support)
    
    # Tìm các cặp phần tử phổ biến đầu tiên
    candidate_itemsets = [(item,) for item in frequent_items]
    frequent_itemsets = []
    while candidate_itemsets:
        frequent_itemsets.extend(candidate_itemsets)
        candidate_itemsets = generate_new_candidates(candidate_itemsets)
        frequent_itemsets = filter_candidates(frequent_itemsets, candidate_itemsets, transactions, min_support)
    
    return frequent_itemsets
    
def generate_new_candidates(itemsets):
    new_itemsets = []
    for i in range(len(itemsets)):
        for j in range(i+1, len(itemsets)):
            itemset1 = itemsets[i]
            itemset2 = itemsets[j]
            if len(itemset1) != len(itemset2):
                continue
            new_itemset = tuple(sorted(set(itemset1) | set(itemset2)))
            if len(new_itemset) == len(itemset1) + 1:
                new_itemsets.append(new_itemset)
    return new_itemsets
    
def filter_candidates(frequent_itemsets, candidate_itemsets, transactions, min_support):
    item_counts = {}
    for transaction in transactions:
        for itemset in candidate_itemsets:
            if set(itemset).issubset(set(transaction)):
                if itemset in item_counts:
                    item_counts[itemset] += 1
                else:
                    item_counts[itemset] = 1
    
    frequent_itemsets = []
    for itemset in candidate_itemsets:
        support = item_counts.get(itemset, 0)/len(transactions)
        if support >= min_support:
            frequent_itemsets.append(itemset)
    
    return frequent_itemsets


transaction_list = [["M1", "M2", "M5"], ["M2","M4"], ["M2","M3"], ["M1", "M2", "M4"],
["M1", "M3"],
["M2", "M3"],
["M1", "M3"],
["M1", "M2", "M3", "M5"],
["M1", "M2", "M3"]]
```
